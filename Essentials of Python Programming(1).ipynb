###Variable-length Arguments


```python
def add(*args):
    sum = 0
    for num in args:
        sum += num
    return sum

print(add(1, 2, 3, 4, 5))
```

```python
def fibonacci(n):
    if n <= 0:
        return "Input should be a positive integer"
    elif n == 1:
        return 0
    elif n == 2:
        return 1
    else:
        a, b = 0, 1
        for _ in range(2, n):
            a, b = b, a + b
        return b

# Example usage:
print(fibonacci(10))  # Output: 34
```

```python
def factorial(n):
    if n < 0:
        return "Input should be a non-negative integer"
    elif n == 0 or n == 1:
        return 1
    else:
        result = 1
        for i in range(2, n + 1):
            result *= i
        return result

# Example usage:
print(factorial(5))  # Output: 120
```

```python
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

# Example usage:
print(gcd(48, 18))  # Output: 6
```

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Example usage:
print(is_prime(29))  # Output: True
print(is_prime(15))  # Output: False
```

#14. Object-Oriented Programming (OOP)
###Creating Classes



```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def greet(self):
        return f"Hello, my name is {self.name} and I am {self.age} years old."

p1 = Person("Alice", 25)
print(p1.greet())
```

###Inheritance

It allows a new class, known as a subclass or derived class, to inherit properties and behaviors (attributes and methods) from an existing class, known as a superclass or base class. This mechanism enables the creation of a hierarchy of classes that share common functionality, reducing code duplication and promoting reusability.

####Key Points of Inheritance
1. Code Reusability: Inheritance allows subclasses to reuse code from the superclass, reducing redundancy and making the codebase easier to maintain.
2. Hierarchical Class Structure: It helps in creating a natural hierarchy of classes, representing relationships like "is-a" between the superclass and subclasses.
3. Method Overriding: Subclasses can override methods from the superclass to provide specific implementations, enabling polymorphism.
4. Extensibility: New functionalities can be added to subclasses without modifying the existing superclass code.

```python
class Student(Person):
    def __init__(self, name, age, student_id):
        super().__init__(name, age)
        self.student_id = student_id

    def get_student_id(self):
        return f"My student ID is {self.student_id}."

s1 = Student("Bob", 20, "S12345")
print(s1.greet())
print(s1.get_student_id())
```

```python
# Base class
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        pass

# Derived class
class Dog(Animal):
    def speak(self):
        return f"{self.name} says Woof!"

# Derived class
class Cat(Animal):
    def speak(self):
        return f"{self.name} says Meow!"

# Creating objects
dog = Dog("Buddy")
cat = Cat("Whiskers")

print(dog.speak())  # Output: Buddy says Woof!
print(cat.speak())  # Output: Whiskers says Meow!
```

###Encapsulation

Encapsulation is one of the fundamental concepts in object-oriented programming (OOP). It refers to the bundling of data (attributes) and methods (functions) that operate on that data into a single unit, typically a class. Encapsulation also involves restricting direct access to some of an object's components, which is a means of preventing unintended interference and misuse of the data. This is typically achieved through the use of access modifiers.

###Key Points of Encapsulation
1. Data Hiding: Encapsulation allows data to be hidden within a class, ensuring that it can only be accessed or modified through specific methods. This is usually done using private or protected access modifiers.
2. Controlled Access: By providing public methods (getters and setters), encapsulation allows controlled access to the internal state of an object. This ensures that the data remains valid and consistent.
3. Modularity: Encapsulation helps in organizing the code into modular and maintainable units. Each class can be developed and tested independently.
4. Security: Encapsulation increases the security of the application by preventing unauthorized access and modification of data.

```python
class BankAccount:
    def __init__(self, balance=0):
        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Insufficient balance!")

    def get_balance(self):
        return self.__balance

account = BankAccount()
account.deposit(100)
account.withdraw(50)
print(account.get_balance())
```

```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.__balance = balance  # Private attribute

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            return self.__balance
        return "Invalid deposit amount"

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            return self.__balance
        return "Invalid withdraw amount"

    def get_balance(self):
        return self.__balance

# Creating an object
account = BankAccount("John")

# Accessing methods
print(account.deposit(100))   # Output: 100
print(account.withdraw(50))   # Output: 50
print(account.get_balance())  # Output: 50

# Trying to access private attribute directly (will cause an error)
# print(account.__balance)   # AttributeError: 'BankAccount' object has no attribute '__balance'
```

###Polymorphism

It allows objects of different classes to be treated as objects of a common superclass. It enables a single interface to be used for different data types, allowing methods to be called on objects of different classes in a uniform way. This is particularly useful for designing systems that are easily extendable and maintainable.

####Key Points of Polymorphism
1. Method Overriding: Allows a subclass to provide a specific implementation of a method that is already defined in its superclass. The method in the subclass overrides the method in the superclass.
2. Method Overloading: (Though not directly supported in Python as it is in some other languages) In general, it allows multiple methods with the same name but different parameters within the same class.
3. Interface-Based Programming: Promotes the use of interfaces or abstract classes to define common methods that different classes should implement.

```python
class Dog:
    def sound(self):
        return "Woof!"

class Cat:
    def sound(self):
        return "Meow!"

def make_sound(animal):
    print(animal.sound())

d = Dog()
c = Cat()

make_sound(d)
make_sound(c)
```

```python
class Shape:
    def area(self):
        pass

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

# List of different shapes
shapes = [Rectangle(2, 3), Circle(5)]

for shape in shapes:
    print(f"The area is {shape.area()}")

# Output:
# The area is 6
# The area is 78.5
```

###Combining Inheritance, Encapsulation, and Polymorphism


```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.__salary = salary  # Encapsulated attribute

    def get_salary(self):
        return self.__salary

    def set_salary(self, salary):
        if salary > 0:
            self.__salary = salary
        else:
            raise ValueError("Salary must be positive")

    def work(self):
        pass

class Developer(Employee):
    def __init__(self, name, salary, programming_language):
        super().__init__(name, salary)
        self.programming_language = programming_language

    def work(self):
        return f"{self.name} is writing code in {self.programming_language}"

class Manager(Employee):
    def __init__(self, name, salary, department):
        super().__init__(name, salary)
        self.department = department

    def work(self):
        return f"{self.name} is managing the {self.department} department"

# Creating objects
dev = Developer("Alice", 100000, "Python")
mgr = Manager("Bob", 120000, "HR")

# Using encapsulated methods
dev.set_salary(105000)
print(dev.get_salary())  # Output: 105000

# Polymorphic behavior
employees = [dev, mgr]
for employee in employees:
    print(employee.work())

# Output:
# Alice is writing code in Python
# Bob is managing the HR department
```

```python
# Base class
class Vehicle:
    def __init__(self, make, model):
        self.make = make
        self.__model = model  # Encapsulated attribute

    def get_model(self):
        return self.__model

    def set_model(self, model):
        if model:
            self.__model = model
        else:
            raise ValueError("Model name cannot be empty")

    def description(self):
        pass

# Derived class
class Car(Vehicle):
    def __init__(self, make, model, num_doors):
        super().__init__(make, model)
        self.num_doors = num_doors

    def description(self):
        return f"{self.make} {self.get_model()} with {self.num_doors} doors"

# Derived class
class Motorcycle(Vehicle):
    def __init__(self, make, model, has_sidecar):
        super().__init__(make, model)
        self.has_sidecar = has_sidecar

    def description(self):
        sidecar_status = "with a sidecar" if self.has_sidecar else "without a sidecar"
        return f"{self.make} {self.get_model()} {sidecar_status}"

# Creating objects
car = Car("Toyota", "Camry", 4)
motorcycle = Motorcycle("Harley-Davidson", "Iron 883", True)

# Using encapsulated methods
car.set_model("Corolla")
print(car.get_model())  # Output: Corolla

# Polymorphic behavior
vehicles = [car, motorcycle]
for vehicle in vehicles:
    print(vehicle.description())

# Output:
# Toyota Corolla with 4 doors
# Harley-Davidson Iron 883 with a sidecar
```

###Key Points:

Inheritance: Car and Motorcycle inherit from Vehicle.

Encapsulation: The __model attribute is private in Vehicle, with getter and setter methods provided.

Polymorphism: The description method is overridden in Car and Motorcycle to provide specific behavior for each type of vehicle

```python
# Base class
class Appliance:
    def __init__(self, brand, power):
        self.brand = brand
        self.__power = power  # Encapsulated attribute

    def get_power(self):
        return self.__power

    def set_power(self, power):
        if power > 0:
            self.__power = power
        else:
            raise ValueError("Power must be a positive value")

    def operate(self):
        pass

# Derived class
class WashingMachine(Appliance):
    def __init__(self, brand, power, load_capacity):
        super().__init__(brand, power)
        self.load_capacity = load_capacity

    def operate(self):
        return f"{self.brand} washing machine with {self.load_capacity}kg load capacity is operating at {self.get_power()} watts."

# Derived class
class Refrigerator(Appliance):
    def __init__(self, brand, power, volume):
        super().__init__(brand, power)
        self.volume = volume

    def operate(self):
        return f"{self.brand} refrigerator with {self.volume} liters volume is operating at {self.get_power()} watts."

# Creating objects
washing_machine = WashingMachine("LG", 2000, 7)
refrigerator = Refrigerator("Samsung", 150, 300)

# Using encapsulated methods
washing_machine.set_power(2100)
print(washing_machine.get_power())  # Output: 2100

# Polymorphic behavior
appliances = [washing_machine, refrigerator]
for appliance in appliances:
    print(appliance.operate())

# Output:
# LG washing machine with 7kg load capacity is operating at 2100 watts.
# Samsung refrigerator with 300 liters volume is operating at 150 watts.
```

###Key Points:
Inheritance: WashingMachine and Refrigerator inherit from Appliance.

Encapsulation: The __power attribute is private in Appliance, with getter and setter methods provided.

Polymorphism: The operate method is overridden in WashingMachine and Refrigerator to provide specific behavior for each type of appliance
