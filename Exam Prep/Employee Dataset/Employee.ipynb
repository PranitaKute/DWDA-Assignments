# Average & Maximun Salary

```python
avg_salary = df['MonthlyIncome'].mean()
print("Average Salary : ",avg_salary)

max_salary = df['MonthlyIncome'].max()
print("Maximum Salary : ", max_salary)
```

# Group By Department & Find Highest Average Salary

```python
grouped = df.groupby('Department')['MonthlyIncome'].mean()
print(grouped)

highest_paid_department = grouped.idxmax()
print("highest_paid_department : ",highest_paid_department)
```
